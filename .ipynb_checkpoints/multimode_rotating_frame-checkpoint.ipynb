Look at a coulpe of modes and a qubit. 


To show approximate (rotating frame) evolution along with the full lab frame numerics one needs to set

    p.show_approx_evolution=True

One can set which terms in the rotating frame are included, by defining their corresponding bessel function indices. So for example to see only the term with bessel function index of 1, one would set:

    p.bessel_indices=[1]

Having fewer terms in the hamiltonian can substantially speed up the calculations, but of course at a cost of accuracy, because we're besiacally neglecting terms that in fact are really there. 


The presence of the counter rotating terms can be controlled via:

    p.non_rwa_terms=1

... to get rid of them, set above to 0. This can speed up the evolution calculations and is implemnted in both the lab as well as the rotated frame.


We set the initial state as the dressed eigenstate, but to start with, "measure" in the bare basis (which commute with the rotating frame transfomration). 



```python
%run multimode.py

#parameters
p=DictExtended(default_sim_parameters)
#p.tlist=np.linspace(0,800,20001)
p.tlist=np.linspace(0,300,100001)

p.non_rwa_terms=0 #1 for keep, 0 for not keep

#static qudit energies
p.q_level_freqs=[
                0.0
                ,5.00 * 2.0*np.pi
                ]

#define functional time dependent portion of the qudit hamiltonian at each energy level
def freq_mod0(t,eps,ome_d): return  0.0*t
def freq_mod1(t,eps,ome_d): return (eps/1.0)*np.sin(ome_d*t)
p.q_level_freqs_time_dep=[
                         freq_mod0
                         ,freq_mod1    
                         ]
p.N_q=len(p.q_level_freqs) #number of levels in a qudit 
p.q_level_decay_rate=[
                (1.0/17000.0)
                ]
p.q_level_dephasing_rate=[
                0.0
                ,(1.0/14000.0)
                ]

p.N_m=3    #umber of levels considered in each mode
p.mode_freqs=[
             6.00 * 2.0 * np.pi
             ,6.35 * 2.0 * np.pi
             ]
p.mode_count=len(p.mode_freqs)
p.mode_q_coupling_freqs=[
                         0.200 *0.3 * 2*np.pi 
                         ,0.200 * 0.45 * 2*np.pi
                        ] 
p.mode_decay_rates=[
                (1.0/5000.0)
                ,(1.0/4000.0)
             ]

p.epsilon=0.300 * 2.0*np.pi


#initial state
p.rho0_name="state_v1"  #1st excited eigenstate of the full system
#p.rho0_name="state_1"  #qubit in 1st excited, rest in ground

#p.states_to_evolve=["state_v0", "state_v1", "state_v2"]  #dressed
p.states_to_evolve=["state_0", "state_1", "state_0_1"] #bare states

p.dissipators_in_dressed_basis=False

p.show_progress_bar=True
p.print_extra_info=True

p.omega_drive=1.01302976  * 2.0*np.pi
#p.omega_drive=(1.01302976  - 0.001) * 2.0*np.pi

#show the rotating frame evolution as well
p.show_approx_evolution=True

#transform the density matrix back to lab frame before calculating expectation values?
p.transform_before_measurement=False

#which terms of the rotated-frame Hamiltonian should we use? in terms fo the bessel function indices.
p.bessel_indices=[1]
#p.bessel_indices=[-2, -1, 0, 1, 2]
#p.bessel_indices=range(-5,6)

ts()
results=run_sim_full_and_approx(p)
print(parameters_to_string(p))
te()
```

So with only one term in the Hamiltonian in the rotating frame the agreement with full evolution is pretty bad!... It's basically because the other time depenent terms contribute... we could fudge things and get the curves to agree much better by applying a dc-offset like correction to the frequency at which we drive at... 

... or let's just include more termsinto the rotating frame Hamiltonian , say onese with bessel function indices -1, 0 and 1:

```python

p.bessel_indices=[-1, 0, 1]

ts()
results=run_sim_full_and_approx(p)
print(parameters_to_string(p))
te()
```

trying to play with less Bessel term

```python

p.bessel_indices=[0, 1]

ts()
results=run_sim_full_and_approx(p)
print(parameters_to_string(p))
te()
```

So even with just a few terms in the rotating frame Hamiltonian, things slow down quite a bit. The wiggles are because we're "measuring" in bare basis, but the full evolution and rotated frame evolution curves are in good agreement. Let's redo the calculation, but now look at the expectation values of dressed basis.

```python

p.states_to_evolve=["state_v0", "state_v1", "state_v2"]  #dressed

ts()
results=run_sim_full_and_approx(p)
print(parameters_to_string(p))
te()
```

In the above plot, the full and rotating frame curves don't (completely) agree - there are some wiggles in the rotating frame evolution. This is because we're measuring dressed basis, which do not commute with the frame transfomation unitary. Redo the calculation but now, before calulating the expectation values, transform the density matrix back into the lab frame. NOTE: THIS WILL BE SLOOOOOOWWWWWW...

```python
#transform the density matrix back to lab frame before calculating expectation values?
p.transform_before_measurement=True

ts()
results=run_sim_full_and_approx(p)
print(parameters_to_string(p))
te()
```

so clearly this the terms nicely overlap now and there are no wiggles, but it's VERY SLOW... It migh be still worthwhile to explore using the rotating frame evolution for doing optimal control, but one would probably not want to tranfrom back into the lab frame... and also probably one would only want to use very few of the terms in the Hamiltonian.
